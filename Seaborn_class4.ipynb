# Seaborn – Class 4: Regression & Mixed Plots

## Regression & Mixed Plots

Seaborn provides plots that combine **relationships** and **statistical models** or mix multiple plot types.

- `regplot()` / `lmplot()` → scatter + regression line  
- `jointplot()` → scatter with hist/KDE  
- `pairplot()` → multiple relationships in one go


# regplot() – Scatter + Regression Line

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
sns.regplot(x=tips['total_bill'], y=tips['tip'])
```

# lmplot() – Regression with Facets

```python
sns.lmplot(x='total_bill',y='tip' ,data=tips)
```

# jointplot() – Scatter + Histogram / KDE

```python
#relation  twn variable
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='reg')
```

# pairplot() – Relationships Between Many Variables

```python
sns.pairplot(tips)
```
