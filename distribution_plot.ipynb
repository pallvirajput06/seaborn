Distribution Plot

```python
# this are created for quantitative variable
# 1.histplot
# 2.jointplot
# 3.pairplot
# 4.runplot
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datasets
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'])
```

```python
#histogram is type of plot which is best resprest distribution of data
#show curnal density astimatter
plt.subplot(1,2,1)
sns.displot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.displot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = "total_bill", y = "tip" , data =df,kind="scatter")
```

```python
sns.jointplot(x = "total_bill", y = "tip" , data =df,kind="hex")
```

```python
sns.jointplot(x = "total_bill", y = "tip" , data =df,kind="kde")
```

```python
sns.jointplot(x = "total_bill", y = "tip" , data =df,kind="reg")
```

Pair ploat

```python
sns.pairplot(df,hue="sex")
#hue is for differnciate
```

```python
sns.pairplot(df,hue="size",palette="husl")
# palette used for coluring in specific way
```
