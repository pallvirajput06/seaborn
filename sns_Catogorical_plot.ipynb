```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('tips')
df
```

```python
df = sns.jointplot(x='total_bill',y='tip',data =df ,kind = 'scatter')
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='kde')
plt.show()
```

```python
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='hex')
plt.show()
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='reg')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue = "sex")
plt.show()
```

COUNT PLOT

```python
df = sns.load_dataset('tips')
sns.countplot(x=df['sex'])
plt.show()
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
plt.show()
```

BAR PLOT

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

default estimator/estimoter

```python
#default estimotor is mean but if i want to change the change like this
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.std)
```

BOX PLOT

```python
sns.boxplot(x='tip',y='day',data=df)
plt.show()
```

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data =df)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data =df)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data =df)
plt.show()
```
